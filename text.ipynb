```python
from sklearn.feature_extraction.text import TfidfVectorizer
text = ['I am bibi', 'you are my bibi', 'you are riding bibi']
tv = TfidfVectorizer()
tv_fit = tv.fit_transform(text)
print(tv.get_feature_names_out())
print(tv.vocabulary_)
print(tv_fit.toarray())
```

```python
import pandas as pd
df = pd.read_csv('Data/news_data.csv', usecols=['time', 'abstract'], index_col=['time'])
trainTexts = df['abstract']
cv0 = CountVectorizer(ngram_range=(2, 2))
tv0 = TfidfVectorizer()
cv0_fit = cv0.fit_transform(trainTexts)
tv0_fit = tv0.fit_transform(trainTexts)
print(cv0.vocabulary_)
print(max(cv0.vocabulary_.values()))
df1 = pd.DataFrame(tv0_fit.todense())
print(df1)
```

```python
from gensim.models import Word2Vec
from nltk.corpus import stopwords
import nltk
import itertools
w2vTexts = df['abstract'].tolist()

w2vTexts = [re.sub(r'([\w])\.([\w])\.', r'\1\2', i) for i in w2vTexts]
w2vTexts = [re.sub('[^\w]+', ' ', i.lower()) for i in w2vTexts]
w2vTexts = [re.sub('[\d]+', '', i) for i in w2vTexts]
w2vTexts = [re.sub('\s+', ' ', i) for i in w2vTexts]

w2vTexts = [nltk.word_tokenize(i) for i in w2vTexts]
w2vTexts = [[i for i in sentence if i not in stopwords.words('english')] for sentence in w2vTexts]

print(w2vTexts)
for i in w2vTexts:
    print(i)
```

```python
custom_embeddings = Word2Vec(w2vTexts, vector_size=50, min_count=2, window=3, epochs=10)
print(custom_embeddings.wv.key_to_index.keys())
print(len(custom_embeddings.wv.key_to_index.keys()))
# print(custom_embeddings.wv['technology'])
print(custom_embeddings.wv.most_similar('predicts', topn=30))
```
