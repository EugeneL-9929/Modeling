```python
import requests
import json
baseUrl = 'https://www.alphavantage.co/query?'
params = {
            'function' : 'TIME_SERIES_INTRADAY',
            '&symbol' : 'IBM',
            '&interval' : '30min',
            '&outputsize' : 'full',
            '&apikey' : 'K7G3VVX3F6EAFJL5',
        }
url = '' + baseUrl
for key, value in params.items():
    if value:
        url += key+'='+value
print(url)
response = requests.get(url).json()
```
